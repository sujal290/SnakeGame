```python
import pdfplumber
import os
import json
from collections import defaultdict

def extract_sgpa_from_pdf(file_path, sem_key):
    data = {}
    try:
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                tables = page.extract_tables()
                if not tables:
                    continue
                for table in tables:
                    for row in table:
                        if not row or row[0] == "Sr.No." or len(row) < 4:
                            continue
                        roll = row[1]
                        name = row[2]
                        sgpa = row[3]
                        if roll and name and sgpa and sgpa.replace(".", "", 1).isdigit():
                            roll = roll.strip()
                            if roll not in data:
                                data[roll] = {"name": name.strip(), "semGPA": {}}
                            data[roll]["semGPA"][sem_key] = float(sgpa.strip())
    except Exception as e:
        print(f"❌ Failed to extract from {file_path}: {e}")
    return data

def collect_all_results(root_folder="results_data"):
    all_results = defaultdict(lambda: {"name": "", "semGPA": {}})
    for sem_dir in sorted(os.listdir(root_folder)):
        sem_path = os.path.join(root_folder, sem_dir)
        if not os.path.isdir(sem_path): continue
        sem_key = sem_dir.lower().replace("sem", "").strip()
        for pdf_file in os.listdir(sem_path):
            if not pdf_file.endswith(".pdf"):
                continue
            file_path = os.path.join(sem_path, pdf_file)
            print(f"📄 Processing: {file_path}")
            results = extract_sgpa_from_pdf(file_path, sem_key)
            for roll, entry in results.items():
                all_results[roll]["name"] = entry["name"]
                all_results[roll]["semGPA"][sem_key] = entry["semGPA"][sem_key]
    return all_results

def compute_cgpa_and_export(all_results, output_path="final_cgpa_results.json"):
    for roll, info in all_results.items():
        gpas = [g for k, g in sorted(info["semGPA"].items(), key=lambda x: int(x[0]))]
        info["semGPA"] = [info["semGPA"].get(str(i), None) for i in range(1, 9)]
        valid_gpas = [g for g in info["semGPA"] if g is not None]
        info["cgpa"] = round(sum(valid_gpas) / len(valid_gpas), 2) if valid_gpas else None
    with open(output_path, "w") as f:
        json.dump(all_results, f, indent=2)
    print(f"\n✅ Extracted {len(all_results)} students. Output saved to {output_path}")
```
